```python
import nest_asyncio
import logging
import csv
import asyncio
from aiogram import Bot, Dispatcher, types, filters
from aiogram.fsm.state import State, StatesGroup
from aiogram.fsm.context import FSMContext
from aiogram.fsm.storage.memory import MemoryStorage
from aiogram.types import InlineKeyboardMarkup, InlineKeyboardButton, CallbackQuery
from aiogram.types import ReplyKeyboardMarkup, KeyboardButton
from aiogram.filters import Command
from aiogram import F
from langchain.schema import HumanMessage, SystemMessage
from langchain.chat_models.gigachat import GigaChat

chat = GigaChat(credentials='YWM1ZGI4MWEtYmU3Mi00ODgwLWFkYmEtYTI2NTk2NjdlYzVhOmVkZGEwYzc3LWVlM2MtNDBhNy1hNGQ3LTg3NDA2NDQ2MzNhZg==', verify_ssl_certs=False)

# Применяем nest_asyncio для поддержки асинхронности в Colab
nest_asyncio.apply()

# Настройка логирования
logging.basicConfig(level=logging.INFO)

# Инициализация бота
API_TOKEN = '7829078479:AAHQaxtM7HfoeFO1I9bQx5IhrYIpCfAMCQY'
bot = Bot(token=API_TOKEN)
dp = Dispatcher(storage=MemoryStorage())

messages = [
    SystemMessage(
        content="Ты бот отправляет строку из № одинаковых элементов, среди которых Х других одинаковых элементов (например строка хххоххх). Пользователь должен внимательно посчитать количество о и ввести ответ в строку. За каждый правильный ответ он получает +1 балл, за каждый неправильный прогресс либо сбрасывается, либо -1 балл (пока не определились). С каждым правильным ответом сложность (количество элементов в строке) увеличивается."
    )
]

# Хранилище зарегистрированных пользователей
registered_users = {}


# Состояния регистрации
class RegistrationStates(StatesGroup):
    waiting_for_name = State()
    waiting_for_email = State()
    filling_survey = State()

# Функция для сохранения данных в CSV
def save_user_data(user_id, name, email):
    with open('users.csv', mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow([user_id, name, email])

# Главное меню
def main_menu():
    return ReplyKeyboardMarkup(
        keyboard=[
            [KeyboardButton(text="Задать вопрос")],
            [KeyboardButton(text="Начать игру")],  # Если нужно другое действие
        ],
        resize_keyboard=True,  # Изменяет размер кнопок под экран устройства
    )

# Кнопка "Начать"
def start_button():
    keyboard = ReplyKeyboardMarkup(
        keyboard=[
            [KeyboardButton(text="Начать")]
        ],
        resize_keyboard=True
    )
    return keyboard

# Обработчик команды /start
@dp.message(Command("start"))
async def start(message: types.Message, state: FSMContext):
    await state.clear()  # Сбрасываем состояние
    user_id = message.from_user.id

    # Если пользователь новый, показываем кнопку "Начать"
    if user_id not in registered_users:
        registered_users[user_id] = {"registered": False, "survey_completed": False}
        await message.answer("Привет! Добро пожаловать! Нажмите 'Начать', чтобы начать регистрацию.", reply_markup=start_button())
    else:
        # Если пользователь уже зарегистрирован, сразу показываем главное меню
        await message.answer("Меню доступно ниже:", reply_markup=main_menu())

# Обработчик нажатия кнопки "Начать"
@dp.message(F.text == "Начать")
async def start_registration(message: types.Message, state: FSMContext):
    user_id = message.from_user.id
    registered_users[user_id]["registered"] = True  # Пользователь начинает регистрацию
    await message.answer("Как вас зовут?")
    await state.set_state(RegistrationStates.waiting_for_name)

@dp.message(F.text == "Начать игру")
async def start_game(message: types.Message):
    question = "сгенерируй строку из N (количество выбреи сам в промежутке от 8 до 20) одинаковых элементов, среди которых Х (количество выбреи сам в промежутке от 1 до 3) других одинаковых элементов (например строка хххоххх). При этом символы, которые нужно посчитать, должны быть случайно распределены внутри строки из 'x' символов. Потом тебе будет отправлен ответ сколько здесь 'o' элементов, и ты должен ответь 'Верно' Или 'Неверно'"
    messages.append(HumanMessage(content=question))
    res = chat(messages)
    messages.append(res)
    await message.answer(res.content)

@dp.message(Command("answer"))
async def start_game(message: types.Message):
    question = message.text[len('/answer '):].strip()
    messages.append(HumanMessage(content=question))
    res = chat(messages)
    messages.append(res)
    await message.answer(res.content)



# Обработчик заполнения имени
@dp.message(filters.StateFilter(RegistrationStates.waiting_for_name))
async def process_name(message: types.Message, state: FSMContext):
    await state.update_data(name=message.text)
    await message.answer("Теперь введите ваш email.")
    await state.set_state(RegistrationStates.waiting_for_email)



# Обработка email с валидацией
@dp.message(filters.StateFilter(RegistrationStates.waiting_for_email))
async def process_email(message: types.Message, state: FSMContext):
    email = message.text
    if "@" not in email:  # Простая проверка на корректность email
        await message.answer("Некорректный email, попробуйте снова.")
        return
    data = await state.get_data()
    name = data["name"]
    user_id = message.from_user.id
    save_user_data(user_id, name, email)

    registered_users[user_id]["survey_completed"] = False  # Анкета не завершена
    await message.answer("Вы успешно зарегистрированы! Заполните анкету.")
    questions = ["Ваш возраст?", "Ваш город?", "Как вы узнали о нас?"]
    await state.update_data(questions=questions, question_index=0, answers=[])
    await state.set_state(RegistrationStates.filling_survey)
    await message.answer(questions[0])

# Обработка вопросов анкеты
@dp.message(filters.StateFilter(RegistrationStates.filling_survey))
async def process_survey_answer(message: types.Message, state: FSMContext):
    data = await state.get_data()
    questions = data["questions"]
    question_index = data["question_index"]

    # Сохраняем ответ
    answers = data.get("answers", [])
    answers.append(message.text)
    await state.update_data(answers=answers)

    next_question_index = question_index + 1
    if next_question_index < len(questions):
        await state.update_data(question_index=next_question_index)
        await message.answer(questions[next_question_index])
    else:
        await message.answer("Анкета завершена! Спасибо за участие.", reply_markup=main_menu())
        registered_users[message.from_user.id]["survey_completed"] = True
        await state.clear()

# Обработчик нажатия кнопки "Задать вопрос"
@dp.message(F.text == "Задать вопрос")
async def ask_question_action(message: types.Message):
    await message.answer("Напишите ваш вопрос после команды /ask.")


# Обработчик команды /ask
@dp.message(Command("ask"))
async def handle_ask_command(message: types.Message, state: FSMContext):
    question = message.text[len('/ask '):].strip()
    messages.append(HumanMessage(content=question))
    res = chat(messages)
    messages.append(res)
    await message.answer(res.content)

# Главное меню
@dp.message(Command("menu"))
async def show_menu(message: types.Message):
    await message.answer("Выберите действие:", reply_markup=main_menu())

async def main():
    try:
        # Убираем webhook, если он был установлен
        await bot.delete_webhook(drop_pending_updates=True)
        # Запуск polling с передачей объекта бота
        await dp.start_polling(bot)
    finally:
        # Закрытие сессии после завершения работы
        await bot.session.close()

if __name__ == "__main__":
    asyncio.run(main())
```
